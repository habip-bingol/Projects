```python
# Hey all !

# This notebook includes a little web scraping proces that includes collect mobile phone's data from n11.com. 
# First of all, I scrapped from only the first page in order to go to step by step 
# Then i added loop in order to the jump other pages
# We can jump the all pages in the website but to show it simply I just collected from 3 pages
```

```python
# import the relevant libraries
import requests
from bs4 import BeautifulSoup
```

```python
# if we don't pass user-agent in header_params parameter, it could occur 403 error (Forbidden) because of the relevant website's services terms and contions. 
# we can find user-agent info from network info in inspect page

url = "https://www.n11.com/telefon-ve-aksesuarlari/cep-telefonu"
header_params = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"}
r = requests.get(url, headers = header_params)
```

```python
r.status_code # controlled the status_code of request and it occurs 200. So we can go on
```

```python
soup = BeautifulSoup(r.content, "html.parser")    # parsing the main page 

# In this step, found all the product's div 
urunler =soup.find_all("div", attrs={"class":"catalogView"})[0].ul.find_all("li", {"class":"column"})


urun_sayisi = 0 # after the finish the loop, it will give  how many product found
for urun in urunler : 
    urun_linki = urun.a.get("href")  # found the link for the each product 
    
    urun_req = requests.get(urun_linki, headers = header_params)   # checked the if was the succesfull
    urun_soup = BeautifulSoup(urun_req.content, "html.parser") # parsing the product's page 
    
    urun_baslik = urun_soup.find("div", attrs={"class":"nameHolder"}).text.strip() 
    
    
    # wrote all label and data in try-except block,  so that not to occur error if scrapper don't find the label-data
    try : 
        urun_label1 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_1"}).find_all("label")[0].text
        urun_data1 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_1"}).find_all("strong")[0].text
    except :
        pass
    
    try : 
        urun_label2 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_2"}).find_all("label")[0].text
        urun_data2 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_2"}).find_all("strong")[0].text
    except :
        pass
    
    try : 
        urun_label3 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_3"}).find_all("label")[0].text
        urun_data3 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_3"}).find_all("strong")[0].text
    except :
        pass
    
    try :
        urun_label4 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_4"}).find_all("label")[0].text
        urun_data4 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_4"}).find_all("strong")[0].text
    except :
        pass
    
    try : 
        urun_label5 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_5"}).find_all("label")[0].text
        urun_data5 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_5"}).find_all("strong")[0].text
    except :
        pass
    
    
   
    fiyat = urun_soup.find_all("div", attrs={"class": "unf-p-summary-price"})[0].text
    
    print("Ürün başlığı  : {}\n\nÜrün özellikleri :\n{:<20} : {}\n{:<20} : {}\n{:<20} : {}\n{:<20} : {}\n\
    {:<20}: {}\n".format(urun_baslik, urun_label1, urun_data1, urun_label2, urun_data2, urun_label3, urun_data3, urun_label4, urun_data4,"Fiyat",fiyat))
    urun_sayisi+=1
    print("*"*30, "\n")

print("Toplamda {} adet ürün bulundu.".format(urun_sayisi))
        
```

```python
url = "https://www.n11.com/telefon-ve-aksesuarlari/cep-telefonu?ipg="
header_params = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"}
```

```python
urun_sayisi = 0
for page in range(1,4):
    r = requests.get(url+str(page), headers = header_params)
    soup = BeautifulSoup(r.content, "html.parser")
    urunler = soup.find_all("div", attrs={"class":"catalogView"})[0].ul.find_all("li", {"class":"column"})
    
    for urun in urunler : 
        urun_linki = urun.a.get("href")

        urun_req = requests.get(urun_linki, headers = header_params)
        urun_soup = BeautifulSoup(urun_req.content, "html.parser")

        urun_baslik = urun_soup.find("div", attrs={"class":"nameHolder"}).text.strip()
    
        try : 
            urun_label1 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_1"}).find_all("label")[0].text
            urun_data1 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_1"}).find_all("strong")[0].text
        except :
            pass
    
        try : 
            urun_label2 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_2"}).find_all("label")[0].text
            urun_data2 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_2"}).find_all("strong")[0].text
        except :
            pass
    
        try : 
            urun_label3 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_3"}).find_all("label")[0].text
            urun_data3 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_3"}).find_all("strong")[0].text
        except :
            pass
    
        try :
            urun_label4 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_4"}).find_all("label")[0].text
            urun_data4 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_4"}).find_all("strong")[0].text
        except :
            pass
    
        try : 
            urun_label5 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_5"}).find_all("label")[0].text
            urun_data5 = urun_soup.find_all("div", attrs={"class":"unf-attribute-cover"})[0].find("div", {"id":"attribute_5"}).find_all("strong")[0].text
        except :
            pass
    
        fiyat = urun_soup.find_all("div", attrs={"class": "unf-p-summary-price"})[0].text
    
        print("Ürün başlığı  : {}\n\nÜrün özellikleri :\n{:<20} : {}\n{:<20} : {}\n{:<20} : {}\n{:<20} : {}\n\
        {:<20}: {}\n".format(urun_baslik, urun_label1, urun_data1, urun_label2, urun_data2, urun_label3, urun_data3, urun_label4, urun_data4,"Fiyat",fiyat))
        
        print("*"*30, "\n")
        urun_sayisi+=1

print("Toplamda {} adet ürün bulundu.".format(urun_sayisi))
        
```
